```python
from sklearn.datasets import load_boston
from scipy import stats
from scipy.stats import zscore
import pandas as pd
import numpy as np
import seaborn as sns
boston = load_boston()
```

```python
data= boston.data
```

```python
cols = boston.feature_names
```

```python
boston_df = pd.DataFrame(data=data, columns=cols)
```

```python
boston_df.head()
```

```python
z = np.abs(boston_df.apply(zscore))
```

```python
boston_df_new = boston_df[(z < 3).any(axis=1)]
```

```python
q1 = boston_df.quantile(0.25)
```

```python
q3 = boston_df.quantile(0.75)
```

```python
IQR = q3-q1
```

```python
cond1 = boston_df < q1 - 1.5*IQR
```

```python
cond2 = boston_df > q3 + 1.5*IQR 
```

```python
boston_df_new[~(cond1 | cond2).any(axis=1) ].shape
```
